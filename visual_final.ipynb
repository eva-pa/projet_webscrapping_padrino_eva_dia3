### Obtenir les données numbeo les plus proches à x mètre du point choisi

```python
res = DistanceFromPoint(df,'lat','lon',-26.20,28.049)
```

```python
res
```

```python
df2 =df.copy(deep=False)
df2['distance_km']=[i/1000 for i in res]
```

```python
df2.sort_values('distance_km')
```

```python
def SortDistancesAscending(df,col_dist):
    return df.sort_values(col_dist)
def DistanceMax(dist_max,sorted_df,col_dist):
    return sorted_df[sorted_df[col_dist]<=dist_max]
```

```python
df3= SortDistancesAscending(df2, 'distance_km')
df3 = DistanceMax(dist_max=100,sorted_df=df3,col_dist='distance_km')
#df3 = DistanceMax(150,'distance_km')
```

```python
df3 = FiltreSortDf(df3,'distance_km',100)
```
