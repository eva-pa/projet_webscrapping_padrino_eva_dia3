```python
import folium
import pandas as pd

# Créer un dictionnaire de données pour les points
"""
data = {'Nom': ['Point 1', 'Point 2', 'Point 3'],
        'Latitude': [45.5236, 45.5236, 45.5236],
        'Longitude': [-122.6750, -122.6750, -122.6750]}
"""
# Créer un dataframe à partir du dictionnaire de données
df = pd.read_csv('C:/Users/evapa/EVA/ECOLE/ESILV/A5/WEBSCRAPPING/projet/projet_webscrapping_padrino_eva_dia3/numbeo/tabAll/tabAllCountries_v2.csv')

# Créer une carte du monde
world_map = folium.Map()

# Pour chaque ligne du dataframe, ajouter un marqueur sur la carte
for i, row in df.iterrows():
    folium.Marker(location=[row['lat'], row['lon']], popup=row['ville']).add_to(world_map)

display(world_map)
```
