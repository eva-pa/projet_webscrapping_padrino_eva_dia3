```python
# Importez les librairies nécessaires
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import pandas as pd

# Ouvrez un navigateur Chrome en utilisant Selenium
driver = webdriver.Chrome("C:/Users/evapa/EVA/ECOLE/ESILV/A5/WEBSCRAPPING/chromium/chromedriver.exe")

# Accédez à la page web
driver.get('https://fr.numbeo.com/co%C3%BBt-de-la-vie/page-de-d%C3%A9marrage')

# Attendez que le menu de sélection de localité soit chargé
wait = WebDriverWait(driver, 10)
location_menu = wait.until(EC.presence_of_element_located((By.ID, "location-menu")))

# Récupérez tous les éléments du menu de sélection de localité
location_items = location_menu.find_elements_by_tag_name("li")

# Créez une liste vide pour stocker les dataframes
dataframes = []

# Pour chaque élément du menu de sélection de localité, cliquez dessus et récupérez le tableau de classement
for location in location_items:
  location.click()

  # Attendez que le tableau soit chargé
  wait.until(EC.presence_of_element_located((By.ID, "ranking-table")))

  # Récupérez le code HTML de la page
  html = driver.page_source

  # Utilisez BeautifulSoup pour parser le code HTML et récupérer le tableau
  soup = BeautifulSoup(html, 'html.parser')
  table = soup.find('table', {'id': 'ranking-table'})

  # Si le tableau existe, créez un dataframe à partir de ses données et ajoutez-le à la liste
  if table:
    df = pd.read_html(str(table))[0]
    dataframes.append(df)

# Fermez le navigateur
driver.quit()

# Affichez les dataframes
print(dataframes)
```
